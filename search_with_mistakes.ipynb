```python
import numpy as np
import math

# вычисление расстояния Левенштейна между 2 словами
def distance(input_str: str, correct_str: str):
    M = len(input_str)
    N = len(correct_str)
    matrix = np.zeros((M + 1, N + 1), dtype = int) # матрица длиной введенного
    for i in range(0, M + 1):
        matrix[i, 0] = i
    for i in range(0, N + 1):
        matrix[0, i] = i

    for i in range(1, M + 1):
        for j in range(1, N + 1):
            if input_str[i - 1] == correct_str[j - 1]:
                cost = 0
            else:
                cost = 1
            matrix[i, j] = min(matrix[i - 1, j - 1] + cost,
                                matrix[i - 1, j] + 1,
                                matrix[i, j - 1] + 1)
    return matrix[M, N]

# получение индекса схожести к введенному слову
def get_index(correct_str, length_input):
    length_correct = len(correct_str)
    if abs(length_correct - length_input) > 19:
        return 0
    else:
        dist = distance(input_str, correct_str)
#    print(correct_str, dist, 1.0 - dist/max (length_input, length_correct))
    return 1.0 - dist/max(length_input, length_correct)

# начало выполнения программы
print('Введите запрос: ')
input_str = input()
input_str = input_str.lower()
length_input = len(input_str)
# if !input_str:
#     return "Вы ничего не ввели!"

for score, out_str in list(sorted(((get_index(str.lower(), length_input), str) for str in current_list), reverse = True))[:10]:
    print(f"{out_str:20}{100*score: 3.2f}")
```
